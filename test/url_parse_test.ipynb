### EuroPMC

```python
keyword="bioinformatics"
pages = 1
perPage = 10

url = f"https://www.ebi.ac.uk/europepmc/webservices/rest/search?query={keyword}&page={pages}&pageSize={perPage}&format=xml"
       
# Get the XML response from Europe PMC.
response = requests.get(url)
tree = ElementTree.fromstring(response.content) # The XML tree object, this variable points to the root
xml_str = response.content.decode('utf-8') if isinstance(response.content, bytes) else response.content # decode the utf-8 encoded response 

# Pretty print
dom = xml.dom.minidom.parseString(xml_str)
pretty_xml = dom.toprettyxml()
print(pretty_xml)

#for result in tree.findall('.//result'):
#    pmid_elem = result.find('pmid')
#    if pmid_elem is not None:
#        pmids.append(pmid_elem.text)
```

### OpenAlex

```python
keyword="bioinformatics"

def get_concept_id(keyword):
    concept_url = f"https://api.openalex.org/concepts?search={keyword}"
    response = requests.get(concept_url)
    content_json = json.loads(response.content)
    # Extract the concept ID
    return content_json['results'][0]['id'].split('/')[3]

get_concept_id(keyword)
```

```python
def get_works(concept_id):
    works_url = f"https://api.openalex.org/works?filter=concepts.id:{concept_id}"
    response = requests.get(works_url)
    content_json = json.loads(response.content)
    
    return content_json

get_works(get_concept_id(keyword))
```
